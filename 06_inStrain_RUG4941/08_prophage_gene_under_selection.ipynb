*How many samples have species which both prophage-carrying and non-carrying population co-exist?

```python
prophage_co_exist_present = prophage_co_exist_microdiversity[["Species", "id"]].drop_duplicates().groupby("id").nunique("Species")
```

```python
os.chdir("/fs/scratch/PAS0439/Ming/virome_ecology_core_prkaryotes/results/08_inStrain_rug4941/instrain_profiles/")
df_list = []
for f in glob.glob("*"):
    i = f.split(".profile")[0]
    path = f"{i}.profile/output/{i}.profile_genome_info.tsv"
    if os.path.exists(path):
        df = pd.read_csv(f"{i}.profile/output/{i}.profile_genome_info.tsv", sep = "\t").query('breadth_minCov >= 0.5')   
    else:
        path = f"{i}.profile/output/{i}.profile_genome_info.tsv.gz"
        df = pd.read_csv(f"{path}", sep = "\t").query('breadth_minCov >= 0.5')
    df.loc[:, "id"] = i
    df_list.append(df)
instrain_genome = pd.concat(df_list)
instrain_genome.genome = [f.split(".fasta")[0] for f in list(instrain_genome.genome)]

instrain_genome_taxa = pd.read_csv("/fs/ess/PAS0439/MING/databases/hungate_rug_rumen_mags_taxa.csv")
instrain_genome_final = pd.merge(instrain_genome, instrain_genome_taxa[["user_genome", "Genera", "Species"]], left_on = "genome", right_on = "user_genome")
```

```python
instrain_genome_present = instrain_genome_final[["Species", "id"]].drop_duplicates().groupby("id").nunique("Species")
```

```python
prophage_co_exist_present.rename(columns = {"Species": "prophage_carrying"}, inplace = True)
```

```python
instrain_genome_present.rename(columns = {"Species": "total"}, inplace = True)
```

```python
prophage_co_exist_present_percent = pd.merge(prophage_co_exist_present, instrain_genome_present, on = "id")
```

```python
for index, row in prophage_co_exist_present_percent.iterrows():
    prophage_co_exist_present_percent.loc[index,"percent"] = prophage_co_exist_present_percent.loc[index,"prophage_carrying"]/prophage_co_exist_present_percent.loc[index,"total"]*100
```
